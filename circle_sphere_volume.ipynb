```python
import numpy as np
import matplotlib.pyplot as plt

# Function to generate random points inside a unit circle
def generate_random_points_in_circle(radius, num_points):
    points = []
    distances = []
    for _ in range(num_points):
        # Generate random radius and angle
        r = radius * np.sqrt(np.random.random())
        theta = np.random.random() * 2 * np.pi
        
        # Convert polar coordinates to Cartesian coordinates
        x = r * np.cos(theta)
        y = r * np.sin(theta)
        
        points.append((x, y))
        distances.append(r)  # Save the radial distance

    return np.array(points), np.array(distances)

# Parameters
radius = 100  # Radius of the circle
num_points = 1000  # Number of random points

# Generate random points
points, distances = generate_random_points_in_circle(radius, num_points)

# Normalize the distances to be between 0 and 1 (for color mapping)
normalized_distances = distances / radius

# Invert the normalized distances for the reversed color effect
inverted_distances = 1 - normalized_distances

# Plot the points with color mapping
plt.figure(figsize=(6, 6))
plt.scatter(points[:, 0], points[:, 1], c=inverted_distances, cmap='viridis', s=5)
plt.gca().set_aspect('equal', adjustable='box')
plt.colorbar(label='Distance from Center (Inverted Normalized)')
plt.title(f'Uniform Random Noise Inside a Circle (Radius = {radius})')
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```

```python
#%matplotlib notebook
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Function to generate random points inside a sphere
def generate_random_points_in_sphere(radius, num_points):
    points = []
    distances = []
    for _ in range(num_points):
        r = radius * np.cbrt(np.random.random())  # Cube root for uniform distribution in volume
        theta = np.random.random() * 2 * np.pi
        phi = np.arccos(2 * np.random.random() - 1)
        
        x = r * np.sin(phi) * np.cos(theta)
        y = r * np.sin(phi) * np.sin(theta)
        z = r * np.cos(phi)
        
        points.append((x, y, z))
        distances.append(r)  # Save the radial distance

    return np.array(points), np.array(distances)

# Parameters
radius = 1  # Radius of the sphere
num_points = 1000  # Number of random points

# Generate random points
points, distances = generate_random_points_in_sphere(radius, num_points)

# Normalize the distances to be between 0 and 1 (for color mapping)
normalized_distances = distances / radius

# Invert the normalized distances for the reversed color effect
inverted_distances = 1 - normalized_distances

# Plot the points with interactivity and color mapping
fig = plt.figure(figsize=(8, 8))
ax = fig.add_subplot(111, projection='3d')

# Use the inverted distances to set the color of each point
scatter = ax.scatter(points[:, 0], points[:, 1], points[:, 2], c=inverted_distances, cmap='viridis', s=2)

# Add a color bar to indicate the distance from the center
cbar = plt.colorbar(scatter, ax=ax)
cbar.set_label('Distance from Center (Inverted Normalized)')

ax.set_aspect('auto')
ax.set_title(f'Uniform Random Noise Inside a Sphere (Radius = {radius})')
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')

plt.show()
```
