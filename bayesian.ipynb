```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

# Step 1: Define the prior distribution for theta (mean house price)
# Assume a normal distribution for the prior with mean 300,000 and standard deviation 50,000
theta_mean = 300000
theta_std = 50000
theta_prior = norm(loc=theta_mean, scale=theta_std)

# Step 2: Define the likelihood function for y given theta
# Assume a normal distribution for house prices given a mean theta with standard deviation 100,000
def likelihood(y, theta):
    return norm.pdf(y, loc=theta, scale=100000)

# Step 3: Compute the prior predictive distribution
# Define the range of house prices we are interested in
y_values = np.linspace(0, 600000, 1000)

# Approximate the integral by summing over many sampled theta values
theta_samples = theta_prior.rvs(size=10000)
prior_predictive = np.zeros_like(y_values)

for theta in theta_samples:
    prior_predictive += likelihood(y_values, theta)

prior_predictive /= len(theta_samples)

# Plot the results
plt.figure(figsize=(10, 6))

# Plot the prior distribution of theta
theta_values = np.linspace(100000, 500000, 1000)
plt.plot(theta_values, theta_prior.pdf(theta_values), label='Prior distribution $f(\\theta)$')

# Plot the prior predictive distribution of y
plt.plot(y_values, prior_predictive, label='Prior predictive distribution $f(y)$')

plt.xlabel('House Price')
plt.ylabel('Density')
plt.legend()
plt.title('Prior and Prior Predictive Distributions')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

# Define the prior distribution for theta (mean house price)
theta_mean = 300000
theta_std = 50000
theta_prior = norm(loc=theta_mean, scale=theta_std)

# Define the likelihood function for y given theta
def likelihood(y, theta):
    return norm.pdf(y, loc=theta, scale=100000)

# Range of house prices
y_values = np.linspace(0, 600000, 1000)
theta_values = np.linspace(theta_mean - 3*theta_std, theta_mean + 3*theta_std, 1000)

# Compute the prior predictive distribution f(y) by integrating over theta
prior_predictive = np.zeros_like(y_values)
theta_prior_pdf = theta_prior.pdf(theta_values)
for i, theta in enumerate(theta_values):
    prior_predictive += likelihood(y_values, theta) * theta_prior_pdf[i]

# Normalize the prior predictive distribution
prior_predictive /= np.trapz(prior_predictive, y_values)  # Use np.trapz for proper normalization

# Plot the results
plt.figure(figsize=(10, 6))
theta_pdf_values = theta_prior.pdf(theta_values)
plt.plot(theta_values, theta_pdf_values, label='Prior distribution $f(\\theta)$')
plt.plot(y_values, prior_predictive, label='Prior predictive distribution $f(y)$')
plt.xlabel('House Price')
plt.ylabel('Density')
plt.legend()
plt.title('Prior and Prior Predictive Distributions')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import expon, norm

# Define the prior distribution for theta (Exponential distribution with rate lambda)
lambda_param = 1/300000
theta_prior = expon(scale=1/lambda_param)

# Define the likelihood function for y given theta
def likelihood(y, theta):
    return norm.pdf(y, loc=theta, scale=100000)

# Determine the range for theta (e.g., up to 5 times the mean)
theta_values = np.linspace(0, 5 * (1/lambda_param), 1000)

# Determine the range for y based on the range of theta
y_min = np.min(theta_values) - 3*100000
y_max = np.max(theta_values) + 3*100000
y_values = np.linspace(y_min, y_max, 1000)

# Compute the prior predictive distribution f(y) by integrating over theta
prior_predictive = np.zeros_like(y_values)
theta_prior_pdf = theta_prior.pdf(theta_values)
for i, theta in enumerate(theta_values):
    prior_predictive += likelihood(y_values, theta) * theta_prior_pdf[i]

# Normalize the prior predictive distribution
prior_predictive /= np.trapz(prior_predictive, y_values)

# Plot the prior distribution of theta
plt.figure(figsize=(10, 6))
theta_pdf_values = theta_prior.pdf(theta_values)
plt.plot(theta_values, theta_pdf_values, label='Prior distribution $f(\\theta)$')

# Plot the prior predictive distribution f(y)
plt.plot(y_values, prior_predictive, label='Prior predictive distribution $f(y)$')
plt.xlabel('House Price')
plt.ylabel('Density')
plt.legend()
plt.title('Prior and Prior Predictive Distributions')
plt.show()
```
