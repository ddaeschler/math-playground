```python
import numpy as np
import matplotlib.pyplot as plt

# Function to draw the unit circle
def plot_unit_circle():
    circle = plt.Circle((0, 0), 1, color='lightblue', fill=False)
    fig, ax = plt.subplots(figsize=(6,6))
    ax.add_artist(circle)
    ax.set_xlim([-1.5, 1.5])
    ax.set_ylim([-1.5, 1.5])
    ax.set_aspect('equal')
    return fig, ax

# Function to visualize the horizontal projection and its square
def plot_cosine_projection_and_square(theta):
    # Calculate cosine value
    cos_theta = np.cos(theta)
    
    # Set up the plot
    fig, ax = plot_unit_circle()
    
    # Plot the point on the unit circle
    x_circle = np.cos(theta)
    y_circle = np.sin(theta)
    ax.plot([0, x_circle], [0, y_circle], color='blue', linestyle='--', label='Radius = 1')
    ax.plot(x_circle, y_circle, 'ro', label=f'Point on circle (cos(θ), sin(θ))')
    
    # Horizontal projection on the x-axis
    ax.plot([x_circle, x_circle], [0, y_circle], color='green', linestyle='--')
    ax.plot([x_circle], [0], 'go', label=f'Projection = cos(θ)')
    
    # Visualizing the square of the projection as an area (cos²(θ))
    # This will be shown as a rectangle with height = cos(θ) and width = cos(θ)
    ax.fill_between([0, cos_theta], 0, cos_theta, color='yellow', alpha=0.3, label=f'cos²(θ) area = {cos_theta**2:.2f}')
    
    # Annotations and labels
    ax.set_title(f'Visualizing cos²(θ) for θ = {np.degrees(theta):.1f}°')
    ax.axhline(0, color='black',linewidth=0.5)
    ax.axvline(0, color='black',linewidth=0.5)
    
    # Label cos(θ) on the x-axis
    ax.text(cos_theta/2, -0.1, f'cos(θ) = {cos_theta:.2f}', horizontalalignment='center', fontsize=12, color='green')
    
    ax.legend()
    plt.show()

# Main function to run the program
if __name__ == "__main__":
    # Input angle θ in degrees and convert it to radians
    theta_deg = float(input("Enter an angle θ (in degrees): "))
    theta_rad = np.radians(theta_deg)
    
    # Plot the cosine projection and its square
    plot_cosine_projection_and_square(theta_rad)
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Function to draw the unit circle
def plot_unit_circle():
    circle = plt.Circle((0, 0), 1, color='lightblue', fill=False)
    fig, ax = plt.subplots(figsize=(6,6))
    ax.add_artist(circle)
    ax.set_xlim([-1.5, 1.5])
    ax.set_ylim([-1.5, 1.5])
    ax.set_aspect('equal')
    return fig, ax

# Function to visualize the horizontal projection, its square, and integration
def plot_cosine_projection_and_integration(start_theta, end_theta, num_steps):
    # Create theta values for integration
    thetas = np.linspace(start_theta, end_theta, num_steps)
    d_theta = thetas[1] - thetas[0]  # Small angle step (d_theta)
    
    # Initialize cumulative area
    cumulative_area = 0
    
    # Set up the plot
    fig, ax = plot_unit_circle()
    
    for theta in thetas:
        # Calculate cosine value and cos²(θ)
        cos_theta = np.cos(theta)
        cos_theta_squared = cos_theta**2
        
        # Compute the area of the small rectangle (width = d_theta, height = cos²(θ))
        area = cos_theta_squared * d_theta
        cumulative_area += area  # Add to cumulative area
        
        # Plot the point on the unit circle
        x_circle = np.cos(theta)
        y_circle = np.sin(theta)
        ax.plot([0, x_circle], [0, y_circle], color='blue', linestyle='--', alpha=0.3)
        
        # Horizontal projection on the x-axis
        ax.plot([x_circle, x_circle], [0, y_circle], color='green', linestyle='--', alpha=0.3)
        ax.plot([x_circle], [0], 'go', alpha=0.3)
        
        # Visualizing the small rectangle (cos²(θ) * d_theta) as an area
        ax.fill_between([theta, theta + d_theta], 0, cos_theta_squared, color='yellow', alpha=0.5)
    
    # Annotations and labels
    ax.set_title(f'Visualizing ∫ cos²(θ) dθ from {np.degrees(start_theta):.1f}° to {np.degrees(end_theta):.1f}°')
    ax.axhline(0, color='black',linewidth=0.5)
    ax.axvline(0, color='black',linewidth=0.5)
    
    ax.legend()
    plt.show()
    
    # Output the total area (simulating the integral)
    print(f"Cumulative area (simulated integral) = {cumulative_area:.4f}")

# Main function to run the program
if __name__ == "__main__":
    # Input the start and end angles in degrees and convert them to radians
    start_theta_deg = float(input("Enter the start angle θ (in degrees): "))
    end_theta_deg = float(input("Enter the end angle θ (in degrees): "))
    num_steps = int(input("Enter the number of small d_θ steps (e.g., 100): "))
    
    start_theta_rad = np.radians(start_theta_deg)
    end_theta_rad = np.radians(end_theta_deg)
    
    # Plot the cosine projection and integration over the range
    plot_cosine_projection_and_integration(start_theta_rad, end_theta_rad, num_steps)
```
